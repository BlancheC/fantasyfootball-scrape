```python
# Imports pandas and reads the csv
import pandas as pd
wrs = pd.read_csv(r'wideRec2021.csv')

```

```python
# Prints the column title of the table 
print(wrs.columns)
```

```python
#Prints the type, 'wrs' is a dataframe
print(type(wrs))
```

```python
#Prints the first five rows of the 'wrs' dataframe
print(wrs.head())
```

```python
# Imports seaborn for analysis

import seaborn as sb
```

```python
# Creates a scatterplot to show relationship
# Shows a negative correlation between Rank and Target

sb.regplot(x='Rank', y='Targets', data=wrs, fit_reg=True).set(title='Scatterplot to show the relationship between Targets and Rank')  
```

```python
# Creates a scatterplot to show relationship
# Shows a negative correlation between Rank and Yards per Reception

sb.regplot(x='Rank', y='Yards per Reception', data=wrs, fit_reg=True).set(title="Scatterplot to show the relationship between Yards per Reception and Rank")
```

```python
# Creates a table showing the correlation between columns in the table

wrs.corr()
```

```python
#Show PairWise Grid, the data shown above as scatterplots
# visualizing the data allows us to notice different patterns. Ex:Games Played vs Rank, top ranking players play more games.

g = sb.PairGrid(wrs)
g.map(sb.scatterplot)
```

```python
# Returns the count, average, standard deviation, minimum, and maximum of all the numeric data
# The average games played per player is 5.5 with 6.3 days per standard deviation.
wrs.describe()
```

```python
# Shows the Histogram of receiving touchdowns per receiver

sb.countplot(x='Receiving Touchdowns', data=wrs).set(title="Distribution of Receiving TDs per Receiver")
```

```python
# Shows the distribution of Receiving touchdowns

sb.displot(wrs, x="Receiving Touchdowns", kind="kde").set(title="Distribution of Receiving TDs per Receiver")
```

```python
import statsmodels.api as sm
```

```python
wrs1=wrs.dropna()
result=sm.OLS(wrs1['Rank'], wrs1['Targets']).fit()
result.summary()
```

```python
result1=sm.OLS(wrs1['Rank'], wrs1['Receptions']).fit()
result1.summary()
```

```python
result2=sm.OLS(wrs1['Rank'], wrs1['Receiving Touchdowns']).fit()
result2.summary()
```
